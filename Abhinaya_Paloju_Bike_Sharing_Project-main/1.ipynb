```python
import pandas as pd

# Example for day.csv
df = pd.read_csv('./data/day.csv', parse_dates=['dteday'])

# Quick info
df.info()
df.describe()
df.isnull().sum()

import matplotlib.pyplot as plt

plt.hist(df['cnt'], bins=30)
plt.title('Distribution of Bike Rentals')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.show()


fig, axes = plt.subplots(2, 2, figsize=(12,8))
axes[0,0].boxplot(df['temp']); axes[0,0].set_title('Temperature')
axes[0,1].boxplot(df['atemp']); axes[0,1].set_title('Feels Like Temp')
axes[1,0].boxplot(df['hum']); axes[1,0].set_title('Humidity')
axes[1,1].boxplot(df['windspeed']); axes[1,1].set_title('Windspeed')
plt.show()



df.groupby(df['dteday'].dt.month)['cnt'].mean().plot(kind='bar', title='Average Rentals per Month')
plt.show()

df.groupby(df['dteday'].dt.weekday)['cnt'].mean().plot(kind='bar', title='Average Rentals by Weekday')
plt.show()

print(df.columns)

# Average rentals by weekday
df.groupby('weekday')['cnt'].mean().plot(kind='bar', title='Average Rentals by Weekday')
plt.show()

df.groupby('mnth')['cnt'].mean().plot(kind='bar', title='Average Rentals by Month')
plt.show()

df = pd.read_csv('./data/hour.csv', parse_dates=['dteday'])
df.groupby('hr')['cnt'].mean().plot(kind='line', title='Average Rentals by Hour')
plt.show()


```
